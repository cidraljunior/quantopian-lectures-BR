##Amostragem

Vamos amostrar dados aleatórios usando uma função do `numpy`.


```python
# Amostra de 100 pontos com média de 0 e desvio padrão de 1. Essa é uma distribuição normal padrão.
X = np.random.normal(0, 1, 100)
```

Podemos usar a biblioteca de visualização de dados que importamos da seguinte maneira.

```python
plt.plot(X)
```

##Ocultar a Linha de Saída

Você deve ter notado a linha de formulário `[<matplotlib.lines.Line2D at 0x7f762c573828>]` antes do gráfico. Isso ocorre porque a função `.plot` realmente produz uma saída. Às vezes, desejamos não exibir a saída, podemos fazer isso com o ponto e vírgula no final.

```python
plt.plot(X);
```

##Adicionando os Títulos dos eixos

Nenhum quant que se preze deixa um gráfico sem eixos rotulados. Aqui estão alguns comandos para ajudar com isso.

```python
X = np.random.normal(0, 1, 100)
X2 = np.random.normal(0, 1, 100)

plt.plot(X);
plt.plot(X2);
plt.xlabel('Tempo') # Os dados que geramos são sem unidade, mas não se esqueça das unidades.
plt.ylabel('Retorno')
plt.legend(['X', 'X2']);
```

##Gerando estatísticas

Vamos usar `numpy`para fazer algumas estatísticas simples.

```python
np.mean(X) # Média
```

```python
np.std(X) # Desvio Padrão
```

##Obtendo Dados Reais de Preços

Dados amostrados aleatoriamente podem ser ótimos para testar idéias, mas vamos obter alguns dados reais. Podemos usar o `get_pricing` para fazer isso. Você pode usar a sintaxe `?` Conforme discutido antes para obter mais informações sobre os argumentos do `get_pricing`.

```python
data = get_pricing('MSFT', start_date='2012-1-1', end_date='2015-6-1')
```

Nossos dados agora são um *dataframe*. Você pode ver o índice *datetime* e as colunas com diferentes dados de preços.

```python
data
```

Esse é um *dataframe* do Pandas, assim podemos indexar para obter apenas os dados de preço.

Para mais informações sobre o Pandas, [aperte aqui](http://pandas.pydata.org/pandas-docs/stable/10min.html).


```python
X = data['price']
```

```python
X.values
```

Como agora também há informações de data nos nossos dados, fornecemos duas series para `.plot`. `X.index` nos fornece o índice de data-hora, e `X.values` nos fornece os valores de preço. Eles são usados como coordenadas X e Y para criar um gráfico.

```python
plt.plot(X.index, X.values)
plt.ylabel('Preço')
plt.legend(['MSFT']);
```

Podemos obter estatísticas sobre dados reais.

```python
np.mean(X) # Média
```

```python
np.std(X) # Desvio Padrão 
```

##Obtendo Retorno dos Preços

Podemos usar a função `pct_change` para obter os retornos. Observe como eliminamos o primeiro elemento depois de fazer isso, pois será `NaN` (nada -> quaquer coisa: resulta em uma alteração percentual de NaN).

```python
R = X.pct_change()[1:]
```

Podemos plotar a distribuição de retornos como um histograma.

```python
plt.hist(R, bins=20)
plt.xlabel('Retorno')
plt.ylabel('Frequência')
plt.legend(['MSFT Retorno']);
```

Obter estatísticas novamente.

```python
np.mean(R) # Média
```

```python
np.std(R) # Desvio Padrão
```

Agora vamos gerar dados a partir de uma distribuição normal usando as estatísticas que estimamos dos retornos da Microsoft. Veremos que temos boas razões para suspeitar que os retornos da Microsoft podem não ser normais, pois a distribuição normal resultante parece muito diferente.


```python
plt.hist(np.random.normal(np.mean(R), np.std(R), 10000), bins=20)
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(['Normally Distributed Returns']);
```

##Gerando média móveis

`pandas`possui boas ferramentas para gerar estatísticas de rolagem. Esse é um exemplo.

Observe como não há média móvel nos primeiros 60 dias, pois não temos 60 dias de dados para gerar a estatística.


```python
# Faz a média dos últimos 30/60/90 dias em cada ponto no tempo.
MAVG30 = X.rolling(window=30,center=False).mean()
MAVG60 = X.rolling(window=60,center=False).mean()
MAVG90 = X.rolling(window=90,center=False).mean()

plt.plot(X.index, X.values)
plt.plot(MAVG30.index, MAVG30.values)
plt.plot(MAVG60.index, MAVG60.values)
plt.plot(MAVG90.index, MAVG90.values)

plt.ylabel('Preço')
plt.legend(['MSFT', '30-day MAVG', '60-day MAVG', '90-day MAVG']);
```
